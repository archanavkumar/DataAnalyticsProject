```python
#response = requests.get('https://api.spoonacular.com/recipes/random?apiKey=b414a4dff99642da8b1ce857fdba9724&addRecipeNutrition=true&number=1')
response = requests.get('https://api.spoonacular.com/recipes/complexSearch?apiKey=b414a4dff99642da8b1ce857fdba9724&addRecipeNutrition=true&number=2')

response.status_code
x = json.loads(response.text)
print (x)
```

```python
recipes_list = []
count = 0
for recipe_info in x['results']:
    recipes_list.append([recipe_info['vegetarian'], recipe_info['glutenFree'], recipe_info['dairyFree'], recipe_info['veryHealthy'], recipe_info['healthScore'], recipe_info['aggregateLikes'], recipe_info['id'], recipe_info['title'], recipe_info['pricePerServing'], recipe_info['readyInMinutes'], recipe_info['servings'], recipe_info['sourceUrl']])
    nutrient_info = []
    for nutrient_info in recipe_info['nutrition']:
        recipes_list.append([recipe_info['nutrition']['nutrients'][count]['name'], recipe_info['nutrition']['nutrients'][count]['amount'], recipe_info['nutrition']['nutrients'][count]['percentOfDailyNeeds']])
        count = count + 1
recipes_list
```

```python
f = open("test.csv", "w+", newline='')
writer = csv.writer(f)

# Write CSV Header
writer.writerow(['vegetarian', 'glutenFree', 'diaryFree', 'veryHealthy', 'healthScore', 'aggregateLikes', 'id', 'title', 'pricePerServing', 'readyInMinutes', 'servings', 'sourceUrl'])
for recipe_info in recipes_list:
    writer.writerow(recipe_info)
f.close()
```
